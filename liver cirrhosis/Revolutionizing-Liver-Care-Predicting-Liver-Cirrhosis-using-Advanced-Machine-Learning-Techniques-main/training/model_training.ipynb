```python
# ğŸ“¦ Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import joblib

# ğŸ§¾ Step 1: Load Data
df = pd.read_csv('liver_data.csv')  # Change filename if needed

# ğŸ§¹ Step 2: Clean and Encode 'gender'
df['gender'] = df['gender'].astype(str).str.strip().str.lower().map({'male': 1, 'female': 0})

# âœ… Check gender encoding
print("âœ… Gender values:", df['gender'].unique())

# ğŸš« Drop rows with any missing values
df.dropna(inplace=True)

# ğŸ¯ Step 3: Define Features & Target
X = df.drop('is_patient', axis=1)
y = df['is_patient']

# ğŸ” Step 4: Check class imbalance
print("\nBefore SMOTE:")
print(y.value_counts())

# ğŸ”„ Step 5: Balance dataset using SMOTE
sm = SMOTE(random_state=42)
X_balanced, y_balanced = sm.fit_resample(X, y)

# âœ… Confirm balance
print("\nAfter SMOTE:")
print(y_balanced.value_counts())

# ğŸ”€ Step 6: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)

# ğŸ“ Step 7: Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ğŸŒ² Step 8: Train the Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# ğŸ“Š Step 9: Evaluate the model
y_pred = model.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)
print(f"\nâœ… Accuracy: {acc * 100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ğŸ’¾ Step 10: Save the model and scaler
joblib.dump(model, 'rf_acc_68.pkl')
joblib.dump(scaler, 'normalizer.pkl')
print("\nğŸ“¦ Model and scaler saved successfully.")
```
